### 2. Evaluation Criteria (1 mark)

The competition uses an **error-based metric** such as **Root Mean Squared Error (RMSE)** or **Mean Absolute Error (MAE)** to evaluate predictions.

#### Root Mean Squared Error (RMSE)
\[
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(\hat{y}_i - y_i)^2}
\]
- **Pros:** Strongly penalises large errors → suitable when big pricing mistakes are costly.  
- **Cons:** Sensitive to outliers (e.g. very expensive cars).

#### Mean Absolute Error (MAE)
\[
MAE = \frac{1}{n}\sum_{i=1}^{n}|\hat{y}_i - y_i|
\]
- **Pros:** More robust to outliers.  
- **Cons:** Does not penalise large deviations as heavily.

Since car prices can vary greatly, **RMSE** is often preferred — it gives higher importance to large pricing errors, which are more costly in the real market.

Extract numerical values (or equivalently eliminate the text) so that the numerical values can be used as a regular feature.

```python
## Task 2, Question 1 Code Here
# === Task 2 — Question 1: Clean numeric-like features; standardise length units to millimetres ===
import pandas as pd
import numpy as np
import re

# 0) Load the raw data
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")

# 1) Configuration
# Columns that contain physical lengths we want in **mm**
LENGTH_COL_KEYS = (
    'back_legroom', 'front_legroom',
    'rear_headroom', 'front_headroom', 'rear_shoulder_room',
    'wheelbase', 'length', 'width', 'height'
)

# Columns we **must not** convert here because Q2 needs their raw text
PROTECT_RAW_TEXT = {
    'vin', 'engine_type', 'torque', 'listed_date',
    'transmission_display', 'wheel_system'
}

_num_pat = re.compile(r'-?\d+(?:\.\d+)?')  # matches integers/decimals, optional negative sign


def mean_if_range(nums, s_lower):
    """
    If the string looks like a range (e.g. '38.5-41 in' or '40 to 42 cm'), average the first two numbers.
    Otherwise return the first number.
    """
    # treat a '-' as range only if it's not a leading negative
    is_range = (' to ' in s_lower) or ('-' in s_lower and not s_lower.strip().startswith('-'))
    if len(nums) >= 2 and is_range:
        return float(np.mean([float(nums[0]), float(nums[1])]))
    return float(nums[0])


def parse_numeric_length_to_mm(x):
    """
    Extract the numeric value from a length string and convert to **millimetres (mm)**.
    Supports: mm, cm, m, inches (" / in / inch / inches). Ranges are averaged.
    Examples:
      '38.5-41 in' -> (38.5+41)/2 inches -> mm
      '102 cm'     -> 1020 mm
    """
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    if not s:
        return np.nan

    s_lower = s.lower().replace(',', '')
    nums = _num_pat.findall(s_lower)
    if not nums:
        return np.nan

    value = mean_if_range(nums, s_lower)

    # Unit detection (order matters; check more specific before generic)
    if 'mm' in s_lower:
        return value  # already millimetres
    if ' cm' in s_lower or s_lower.endswith('cm'):
        return value * 10.0  # centimetres -> mm
    # ' m ' (metres) but not 'mm'/'cm'
    if re.search(r'(^|\W)m(\W|$)', s_lower) and not re.search(r'mm|cm', s_lower):
        return value * 1000.0  # metres -> mm
    if any(u in s_lower for u in ['"', ' in', ' inch', ' inches']):
        return value * 25.4  # inches -> mm

    # Default: if unit not shown, many car datasets mean inches for legroom/headroom; convert to mm
    return value * 25.4


def extract_numeric_generic(x):
    """
    Generic numeric extractor for columns like price, power, etc.
    Takes the first number (or mean of first two in a range); strips commas.
    """
    if pd.isna(x):
        return np.nan
    s = str(x).strip().lower().replace(',', '')
    nums = _num_pat.findall(s)
    if not nums:
        return np.nan
    return mean_if_range(nums, s)


def clean_all_numeric_like_columns(
    df: pd.DataFrame,
    treat_as_length_cols=LENGTH_COL_KEYS,
    protect_cols=PROTECT_RAW_TEXT,
    overwrite=True
) -> pd.DataFrame:
    """
    Convert object/text columns that contain numeric information into float columns.
    - Any column whose name matches a length key is converted to **mm**.
    - Other numeric-like object columns have their numeric values extracted.
    - Columns listed in `protect_cols` are **skipped** to preserve raw text for Q2.
    Returns a small DataFrame summarising what was converted (for evidence).
    """
    summary = []
    obj_cols = [c for c in df.columns if df[c].dtype == 'O']

    for col in obj_cols:
        # keep raw text for the protected columns
        if col in protect_cols:
            continue

        ser = df[col].astype(str)
        non_blank = ser[ser.str.strip().ne('')]
        if non_blank.empty:
            continue

        # is this column "numeric-like" (has digits) in most rows?
        has_number = non_blank.str.contains(_num_pat)
        if has_number.mean() < 0.60:
            continue  # skip columns that are mostly non-numeric text

        is_length = any(k == col.lower() or k in col.lower() for k in treat_as_length_cols)
        parser = parse_numeric_length_to_mm if is_length else extract_numeric_generic

        cleaned_vals = non_blank.map(parser)

        # align back to full index
        col_clean = pd.Series(np.nan, index=df.index, dtype='float64')
        col_clean.loc[non_blank.index] = cleaned_vals

        target = col if overwrite else f"{col}_num"
        df[target] = col_clean

        summary.append({
            "column": col,
            "target": target,
            "converted_non_null_rows": int(col_clean.notna().sum()),
            "null_after": int(col_clean.isna().sum())
        })

    return pd.DataFrame(summary)


# 2) Run the cleaner
train_q1_report = clean_all_numeric_like_columns(train, overwrite=True)
test_q1_report  = clean_all_numeric_like_columns(test,  overwrite=True)

# 3) Rename length columns to make the **unit explicit** (_mm)
def rename_mm(df):
    for col in LENGTH_COL_KEYS:
        if col in df.columns and pd.api.types.is_float_dtype(df[col]):
            df.rename(columns={col: f"{col}_mm"}, inplace=True)

rename_mm(train)
rename_mm(test)

# 4) Save cleaned datasets + conversion reports (evidence)
train.to_csv("train_q1mm_clean.csv", index=False)
test.to_csv("test_q1mm_clean.csv", index=False)
train_q1_report.to_csv("train_q1mm_report.csv", index=False)
test_q1_report.to_csv("test_q1mm_report.csv", index=False)

print("Q1 finished — saved: train_q1mm_clean.csv, test_q1mm_clean.csv, and the *_report.csv files.")

# 5) Helpful evidence prints for your notebook
#    (a) Show which columns were converted
from IPython.display import display

print("\nQ1 — Converted columns (TRAIN):")
display(train_q1_report.head(20))

print("\nQ1 — Converted columns (TEST):")
display(test_q1_report.head(20))

#    (b) Dtypes AFTER for key length columns (should be float64 with _mm suffix)
length_mm_cols = [c for c in ['back_legroom_mm','front_legroom_mm','wheelbase_mm','width_mm','height_mm',
                              'rear_headroom_mm','front_headroom_mm','rear_shoulder_room_mm'] if c in train.columns]
print("\nQ1 — Dtypes AFTER for key length columns (TRAIN):")
for c in length_mm_cols:
    print(f"  {c}: {train[c].dtype}")

#    (c) Before vs After sample for a few length columns (first 5 rows)
raw_cols  = [c for c in ['back_legroom','front_legroom','wheelbase','width','height'] if c in pd.read_csv('train.csv').columns]
after_cols = [c for c in ['back_legroom_mm','front_legroom_mm','wheelbase_mm','width_mm','height_mm'] if c in train.columns]

if raw_cols and after_cols:
    print("\nQ1 — Before (raw) vs After (mm) sample (first 5 rows):")
    raw_preview = pd.read_csv('train.csv')[raw_cols].head(5).add_prefix('before_')
    after_preview = train[after_cols].head(5).add_prefix('after_')
    display(pd.concat([raw_preview, after_preview], axis=1))
else:
    print("\nQ1 — Note: could not find both raw and cleaned length columns to align for preview.")
```

```python
## Task 2, Question 2 Code Here


import os, re, numpy as np, pandas as pd

# 0) Delete old Q2 CSVs if they exist (prevents confusion)
for f in ["train_q2mm_features.csv", "test_q2mm_features.csv"]:
    try:
        os.remove(f)
        print(f"Deleted old file: {f}")
    except FileNotFoundError:
        pass

# 1) Load Q1-clean data (length columns already in mm)
train_q1 = pd.read_csv("train_q1mm_clean.csv")
test_q1  = pd.read_csv("test_q1mm_clean.csv")

# 2) Helpers to locate columns even if names differ a bit
def _first_present(df, names):
    """Return the first column in df whose lowercase name matches/contains any in `names`."""
    names = [n.lower() for n in names]
    low_map = {c.lower(): c for c in df.columns}
    # exact first
    for n in names:
        if n in low_map: 
            return low_map[n]
    # fuzzy contains
    for c in df.columns:
        cl = c.lower()
        if any(n in cl for n in names):
            return c
    return None

# 3) Robust feature engineering
def engineer_features(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    # --- locate likely source columns ---
    torque_col = _first_present(out, ['torque','torque_lbft','torque_nm'])
    trans_col  = _first_present(out, ['transmission_display','transmission','gearbox'])
    eng_col    = _first_present(out, ['engine_type','engine','engine_description'])
    drive_col  = _first_present(out, ['wheel_system','drivetrain','drive','drive_type'])
    date_col   = _first_present(out, ['listed_date','listing_date','listed_on','date_listed','created_at'])

    # --- torque (lb-ft) & rpm (supports Nm -> lb-ft conversion) ---
    def parse_torque_lbft(s):
        if pd.isna(s): return np.nan
        t = str(s).lower()
        m_nm = re.search(r'(\d+(?:\.\d+)?)\s*nm\b', t)
        if m_nm:  # convert Nm to lb-ft
            return float(m_nm.group(1)) * 0.737562
        m_lb = re.search(r'(\d+(?:\.\d+)?)\s*(?:lb[-\s]?ft|lbft|ft[-\s]?lb|ft[-\s]?lbs?)\b', t)
        return float(m_lb.group(1)) if m_lb else np.nan

    def parse_torque_rpm(s):
        if pd.isna(s): return np.nan
        t = str(s).lower().replace(',', '')
        m = re.search(r'(?:@|at)\s*(\d+)\s*(?:rpm|rev/?min)\b', t)
        return float(m.group(1)) if m else np.nan

    out['torque_lbft'] = out[torque_col].map(parse_torque_lbft) if torque_col else np.nan
    out['torque_rpm']  = out[torque_col].map(parse_torque_rpm)  if torque_col else np.nan

    # --- transmission: speeds & automatic flag ---
    def trans_speeds(s):
        if pd.isna(s): return np.nan
        t = str(s).lower()
        m = re.search(r'(\d+)\s*[- ]*(?:speed|spd)\b', t)
        return float(m.group(1)) if m else np.nan

    def trans_is_auto(s):
        if pd.isna(s): return np.nan
        t = str(s).lower()
        if any(k in t for k in ['automatic','cvt','continuously variable','dual-clutch','dct','dsg','amt','a/t','auto ']): return 1
        if any(k in t for k in ['manual','m/t']): return 0
        return np.nan

    if trans_col:
        out['trans_speeds']  = out[trans_col].map(trans_speeds)
        out['trans_is_auto'] = out[trans_col].map(trans_is_auto).astype('Float64')
    else:
        out['trans_speeds']  = np.nan
        out['trans_is_auto'] = pd.Series(np.nan, index=out.index, dtype='Float64')

    # --- engine: cylinders & layout (V / I / W / FLAT) ---
    def parse_engine_cyl(s):
        if pd.isna(s): return np.nan
        m = re.search(r'(\d+)', str(s))
        return float(m.group(1)) if m else np.nan

    def parse_engine_layout(s):
        if pd.isna(s): return np.nan
        t = str(s).upper()
        if re.search(r'\bV\s*[-]?\s*\d', t): return 'V'
        if re.search(r'\bI\s*[-]?\s*\d|\bINLINE', t): return 'I'
        if re.search(r'\bW\s*[-]?\s*\d', t): return 'W'
        if any(k in t for k in ['FLAT','BOXER','HORIZONTALLY OPPOSED']): return 'FLAT'
        return np.nan

    if eng_col:
        out['engine_cylinders'] = out[eng_col].map(parse_engine_cyl)
        out['engine_layout']    = out[eng_col].map(parse_engine_layout)
    else:
        out['engine_cylinders'] = np.nan
        out['engine_layout']    = np.nan

    # --- drivetrain AWD/4WD flag ---
    def is_awd4wd(s):
        t = str(s).upper()
        return 1 if any(k in t for k in ['AWD','4WD','4X4']) else 0

    if drive_col:
        out['is_awd4wd'] = out[drive_col].map(is_awd4wd).astype('Int64')
    else:
        out['is_awd4wd'] = pd.Series(0, index=out.index, dtype='Int64')  # safe default

    # --- listed date: handle text formats + epoch seconds/milliseconds ---
    if date_col:
        raw = out[date_col].astype(str).str.strip().replace({'0': np.nan, '0000-00-00': np.nan, '': np.nan})
        is_numeric = raw.str.fullmatch(r'\d+').fillna(False)
        parsed = pd.Series(pd.NaT, index=out.index, dtype='datetime64[ns]')

        if is_numeric.any():
            nums = pd.to_numeric(raw.where(is_numeric), errors='coerce')
            # Heuristic: >=12 digits → milliseconds; else seconds
            ms_mask = nums.astype('Int64').astype(str).str.len() >= 12
            parsed.loc[is_numeric & ms_mask]  = pd.to_datetime(nums[is_numeric & ms_mask],  unit='ms', errors='coerce')
            parsed.loc[is_numeric & ~ms_mask] = pd.to_datetime(nums[is_numeric & ~ms_mask], unit='s',  errors='coerce')

        parsed.loc[~is_numeric] = pd.to_datetime(raw[~is_numeric], errors='coerce', format=None)
        out['listed_date'] = parsed

        # treat epoch artefacts as missing if still 1970
        out.loc[out['listed_date'].dt.year == 1970, 'listed_date'] = pd.NaT
        out['listed_year']  = out['listed_date'].dt.year
        out['listed_month'] = out['listed_date'].dt.month
    else:
        out['listed_year']  = np.nan
        out['listed_month'] = np.nan

    # --- vehicle age (listed_year - model year) ---
    if 'year' in out.columns:
        out['vehicle_age'] = out['listed_year'] - out['year']
    else:
        out['vehicle_age'] = np.nan

    return out

# 4) Build features
train_q2 = engineer_features(train_q1)
test_q2  = engineer_features(test_q1)

# 5) Evidence for the marker
new_cols = [c for c in train_q2.columns if c not in train_q1.columns]
print("Q2 — New feature columns:")
print(new_cols)

print("\nQ2 — Non-null counts (TRAIN):")
display(train_q2[new_cols].notna().sum().to_frame("non_null"))

print("\nQ2 — Preview (first 5 rows):")
display(train_q2[new_cols].head(5))

# 6) Save fresh Q2 outputs (overwriting is fine; we deleted old ones above)
train_q2.to_csv("train_q2mm_features.csv", index=False)
test_q2.to_csv("test_q2mm_features.csv", index=False)
print("\nSaved: train_q2mm_features.csv, test_q2mm_features.csv")

```

```python
## Task 2, Question 4 Code Here
import pandas as pd 
import numpy as np 

train = pd.read_csv("train_q3mm_imputed.csv")
test  = pd.read_csv("test_q3mm_imputed.csv")

target_col = "price" if "price" in train.columns else None
X_train = train.drop(columns=[target_col]) if target_col else train.copy()
y_train = train[target_col] if target_col else None
X_test  = test.copy()

num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_train.columns if c not in num_cols]

def cap_categories_to_top_k(s: pd.Series, k=5):
    counts = s.value_counts()
    topk = counts.index[:k]
    return s.where(s.isin(topk), "other"), set(topk)

for c in cat_cols:
    X_train[c], topk = cap_categories_to_top_k(X_train[c].astype(str), k=5)
    X_test[c] = X_test[c].astype(str).where(X_test[c].astype(str).isin(topk), "other")

X_train_enc = pd.get_dummies(X_train, columns=cat_cols, drop_first=False)
X_test_enc  = pd.get_dummies(X_test,  columns=cat_cols, drop_first=False)
X_test_enc  = X_test_enc.reindex(columns=X_train_enc.columns, fill_value=0)

train_q4 = pd.concat([X_train_enc, y_train], axis=1) if target_col else X_train_enc
test_q4  = X_test_enc

train_q4.to_csv("train_q4mm_encoded.csv", index=False)
test_q4.to_csv("test_q4mm_encoded.csv",  index=False)
```

## Question 2

- Multiple Linear Regression (OLS): Fast, simple, and interpretable, useful as a baseline because it shows the direct, average effect of each feature on price, so you can quickly see what matters and sanity-check the data.

- Ridge Regression (L2): Improves linear accuracy and stability because the L2 penalty shrinks coefficients for correlated predictors (e.g., horsepower/torque/power; year/age), reducing overfitting and variance while staying interpretable.

- Random Forest: Learns complex patterns because it naturally captures nonlinearities and interactions (e.g., diminishing returns at high horsepower, mileage × year effects) and is robust to outliers/scale, often yielding stronger predictive performance on tabular data.

```python
# Question 3
import pandas as pd, numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import TransformedTargetRegressor

# --- load your data (same as before) ---
TRAIN = "train_q2mm_features.csv"
TARGET = "price"
ID_COL = "vin"

train = pd.read_csv(TRAIN)
X_all = train.drop(columns=[c for c in [TARGET, ID_COL] if c in train.columns])
NUM_COLS = X_all.select_dtypes(include=[np.number]).columns.tolist()
X = train[NUM_COLS].copy()
y = train[TARGET].astype(float).values

cv = KFold(n_splits=5, shuffle=True, random_state=42)

# 1) OLS (with imputer!)
ols_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("model", LinearRegression())
])

ols_rmse = -cross_val_score(
    ols_pipe, X, y, cv=cv,
    scoring="neg_root_mean_squared_error"
)
print(f"OLS CV RMSE: mean={ols_rmse.mean():.0f}, std={ols_rmse.std():.0f}")

# 2) Ridge (with imputer!)
ridge_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("model", Ridge(alpha=10.0))
])

ridge_rmse = -cross_val_score(
    ridge_pipe, X, y, cv=cv,
    scoring="neg_root_mean_squared_error"
)
print(f"Ridge (alpha=10) CV RMSE: mean={ridge_rmse.mean():.0f}, std={ridge_rmse.std():.0f}")

# 3) Random Forest on log-price (with imputer!) -- your best
rf_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("rf", RandomForestRegressor(
        n_estimators=1000, max_depth=24,
        min_samples_leaf=2, max_features=0.6,
        bootstrap=True, n_jobs=-1, random_state=42
    ))
])

rf_log = TransformedTargetRegressor(
    regressor=rf_pipe,
    func=np.log1p,
    inverse_func=np.expm1
)

rf_rmsle = -cross_val_score(
    rf_log, X, y, cv=cv,
    scoring="neg_mean_squared_log_error"
)
print(f"RF(log target) CV RMSLE: mean={np.sqrt(rf_rmsle.mean()):.5f}")
```

1) Multiple Linear Regression (OLS)

- Fitted weights (direction): Positive for performance specs (horsepower/torque/power, trans_speeds, dimensions), negative for usage/age (mileage, vehicle_age). This matches the EDA (higher performance increased price; higher mileage/vehicle_age decreased price.).

- Hyperparameters: None.

- Predictive performance: CV RMSE ≈ 9,555 (+-642). The relatively large fold-to-fold spread and known collinearity (e.g., horsepower–torque–power; year–age) make OLS coefficients less stable.

2) Ridge Regression (L2)

- Fitted weights (effect): Same signs as OLS but shrunk and more stable across correlated groups (Ridge shares weight among horsepower/torque/power and between year/vehicle_age).

- Optimised hyperparameter: alpha = 10.0 (log-grid search).

- Predictive performance: CV RMSE = 9,554, essentially the same as OLS here, indicating linear structure alone is insufficient; the gain comes mainly from non-linear modelling rather than just regularising coefficients.

3) Random Forest Regressor

- Feature effects: Captures the nonlinear patterns seen in scatterplots (diminishing returns at high horsepower; nonlinear depreciation with mileage) and interactions (e.g., mileage × year/segment). In practice, importance is typically led by horsepower/mileage, followed by year/vehicle_age and key specs like torque/power/trans_speeds.

- Optimised hyperparameters: max_depth=24, max_features=0.6, min_samples_leaf=1, n_estimators=400.

- Predictive performance: CV RMSE ≈ 5,473, which is ~43% lower than Ridge/OLS ((9554 - 5473 ) / 9554 = 0.4). This sizeable improvement indicates meaningful nonlinearity/interaction that linear models miss.

```python
# Question 4
import pandas as pd, numpy as np
from pathlib import Path
from itertools import product
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import TransformedTargetRegressor
from sklearn.ensemble import RandomForestRegressor

# --- config
TRAIN, TEST, SAMPLE, TARGET, ID_COL = "train_q2mm_features.csv", "test_q2mm_features.csv", "sample_solution.csv", "price", "vin"
SEED = 42
cv = KFold(n_splits=5, shuffle=True, random_state=SEED)  # reproducible CV

# --- load
train = pd.read_csv(TRAIN, dtype={ID_COL: str})
test  = pd.read_csv(TEST,  dtype={ID_COL: str})
use_sample = Path(SAMPLE).exists()
sample = pd.read_csv(SAMPLE, dtype={ID_COL: str}) if use_sample else None

# --- features: SAME numeric columns as Task-2
X_all = train.drop(columns=[c for c in [TARGET, ID_COL] if c in train.columns])
NUM_COLS = X_all.select_dtypes(include=[np.number]).columns.tolist()
if not NUM_COLS:
    raise ValueError("No numeric Task-2 features found.")
X = train[NUM_COLS].copy()
y = train[TARGET].astype(float).values

# align test to VIN order (prefer sample order)
if use_sample:
    test_aligned = sample[[ID_COL]].merge(test[[ID_COL] + NUM_COLS], on=ID_COL, how="left")
    if len(test_aligned) != len(sample):
        raise ValueError("VIN mismatch between sample_solution.csv and test.")
    X_test = test_aligned[NUM_COLS].values
    VIN_OUT = test_aligned[ID_COL].astype(str).values
else:
    X_test = test[NUM_COLS].values
    VIN_OUT = test[ID_COL].astype(str).values

# --- small, safe RF grid around your best
grid = dict(
    n_estimators=[800, 1000, 1500],
    max_depth=[20, 24, 28],
    min_samples_leaf=[1, 2, 3],
    max_features=[0.6, "sqrt"],
)
candidates = list(product(*grid.values()))
keys = list(grid.keys())

def make_est(params, seed=SEED):
    kw = dict(zip(keys, params))
    rf = Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("rf",  RandomForestRegressor(
            random_state=seed, n_jobs=-1, bootstrap=True, **kw
        ))
    ])
    # model log1p(price) and invert with expm1
    return TransformedTargetRegressor(regressor=rf, func=np.log1p, inverse_func=np.expm1)

# --- CV search (RMSLE)
best_params, best_cv = None, np.inf
for p in candidates:
    est = make_est(p, seed=SEED)
    # CV RMSLE = sqrt(MSLE). Use neg_mean_squared_log_error and sqrt the mean.
    msle = -cross_val_score(est, X, y, cv=cv, scoring="neg_mean_squared_log_error", n_jobs=-1)
    cv_rmsle = float(np.sqrt(msle.mean()))
    if cv_rmsle < best_cv:
        best_cv, best_params = cv_rmsle, p

print("Best CV RMSLE:", round(best_cv, 6))
print("Best params:", dict(zip(keys, best_params)))

# --- fit best on all training data (single model)
best_model = make_est(best_params, seed=SEED)
best_model.fit(X, y)
preds_single = np.maximum(best_model.predict(X_test), 0.0)

# write single-model submission
out_single = pd.DataFrame({"vin": VIN_OUT, "price": preds_single.astype(float)})
assert list(out_single.columns) == ["vin", "price"]
assert out_single["vin"].is_unique and out_single["price"].notna().all()
out_single.to_csv("submission_rf_log_cvbest.csv", index=False)
print("Wrote submission_rf_log_cvbest.csv  rows:", len(out_single), "  features:", len(NUM_COLS))

# --- seed-bagging (same best hyperparams, different random_state)
SEED_BAG = [11, 22, 33, 44, 55]
bag_preds = []
for s in SEED_BAG:
    m = make_est(best_params, seed=s)
    m.fit(X, y)
    bag_preds.append(m.predict(X_test))
preds_bag = np.maximum(np.mean(bag_preds, axis=0), 0.0)

# write bagged submission
out_bag = pd.DataFrame({"vin": VIN_OUT, "price": preds_bag.astype(float)})
assert list(out_bag.columns) == ["vin", "price"]
assert out_bag["vin"].is_unique and out_bag["price"].notna().all()
out_bag.to_csv("submission_rf_log_cvbest_bag5.csv", index=False)
print("Wrote submission_rf_log_cvbest_bag5.csv  rows:", len(out_bag), "  features:", len(NUM_COLS))
```

**Comment**
- I used a Random Forest on log-price (median-impute → RF, predicting log1p(price) then back-transforming) because EDA showed right-skewed prices and curved/interactive effects (e.g., horsepower with age/mileage) that linear models underfit; RF handles nonlinearity and collinearity without new features, while the log transform stabilizes variance and aligns with RMSLE-style evaluation. A tight 5-fold CV grid around the RF baseline (trees≈1000–1500, depth≈20–28, leaf=2–3, max_features 0.6/“sqrt”, bootstrap) selected the final settings, and a small seed-bag reduced variance. On Kaggle’s public leaderboard (50% test), my best submission scored 0.10307, improving steadily from earlier runs (0.1206 > 0.1084 > 0.1031); this matches CV performance, suggesting good generalisation and avoiding public-split overfitting ahead of the final ranking.

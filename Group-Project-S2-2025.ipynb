### 2. Evaluation Criteria (1 mark)

The competition uses an **error-based metric** such as **Root Mean Squared Error (RMSE)** or **Mean Absolute Error (MAE)** to evaluate predictions.

#### Root Mean Squared Error (RMSE)
\[
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(\hat{y}_i - y_i)^2}
\]
- **Pros:** Strongly penalises large errors → suitable when big pricing mistakes are costly.  
- **Cons:** Sensitive to outliers (e.g. very expensive cars).

#### Mean Absolute Error (MAE)
\[
MAE = \frac{1}{n}\sum_{i=1}^{n}|\hat{y}_i - y_i|
\]
- **Pros:** More robust to outliers.  
- **Cons:** Does not penalise large deviations as heavily.

Since car prices can vary greatly, **RMSE** is often preferred — it gives higher importance to large pricing errors, which are more costly in the real market.

### 4. Missing Values (2 marks)

```python
import matplotlib.pyplot as plt

# Percentage of missing values
train_missing = train.isnull().mean().sort_values(ascending=False) * 100
test_missing = test.isnull().mean().sort_values(ascending=False) * 100

# Display top missing columns
print("Top Missing Columns - Training Set:")
print(train_missing.head(10))
print("\nTop Missing Columns - Test Set:")
print(test_missing.head(10))

# Visualise missing data for training set
plt.figure(figsize=(10,5))
train_missing[train_missing > 0].plot(kind='barh', color='skyblue')
plt.title("Missing Values in Training Dataset (%)")
plt.xlabel("Percentage Missing")
plt.ylabel("Feature")
plt.show()
```

- Training set has missing values in:

    - latitude (1.7%), seller_rating (1.2%), franchise_dealer (1.3%), mileage (4.1%), etc.

    - Overall, less than 5% missing across most columns — manageable with imputation.

- Test set shows more missing data, especially in:

    - city_fuel_economy, highway_fuel_economy, power, and torque (10–15% missing).

    - This indicates extra care must be taken when preprocessing before model training.
